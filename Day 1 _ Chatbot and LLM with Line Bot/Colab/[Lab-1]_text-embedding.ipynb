```python
# prompt: Calculate cosine similarity of text with NumPy

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def calculate_cosine_similarity(text1_embedding, text2_embedding):
    # Reshape the embeddings to be 2D arrays for cosine_similarity
    text1_embedding = np.array(text1_embedding).reshape(1, -1)
    text2_embedding = np.array(text2_embedding).reshape(1, -1)

    similarity = cosine_similarity(text1_embedding, text2_embedding)[0][0]
    return similarity

# Example usage (assuming you have embeddings for text1, text2, etc.)
similarity_1_2 = calculate_cosine_similarity(embeddings, embeddings2)
similarity_1_text1 = calculate_cosine_similarity(embeddings, embed_model.get_text_embedding(text1))
similarity_text1_text2 = calculate_cosine_similarity(embed_model.get_text_embedding(text1), embed_model.get_text_embedding(text2))
similarity_text1_text3 = calculate_cosine_similarity(embed_model.get_text_embedding(text1), embed_model.get_text_embedding(text3))
similarity_text1_text4 = calculate_cosine_similarity(embed_model.get_text_embedding(text1), embed_model.get_text_embedding(text4))


print(f"Cosine Similarity between embeddings and embeddings2: {similarity_1_2}")
print(f"Cosine Similarity between embeddings and text1: {similarity_1_text1}")
print(f"Cosine Similarity between text1 and text2: {similarity_text1_text2}")
print(f"Cosine Similarity between text1 and text3: {similarity_text1_text3}")
print(f"Cosine Similarity between text1 and text4: {similarity_text1_text4}")
```

The TokenTextSplitter attempts to split to a consistent chunk size according to raw token counts.

```python
from llama_index.core.node_parser import TokenTextSplitter
splitter = TokenTextSplitter(
    chunk_size=100,
    chunk_overlap=50,
    separator=" ",
)
```

```python
from llama_index.core import Document
```

```python
nodes = splitter.get_nodes_from_documents(
    [Document(text=text1+text2+text3+text4)], show_progress=True
)
```
